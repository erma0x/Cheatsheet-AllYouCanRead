# Pipenv devops tutorial

### 1. go in the folder of your project
cd my_project/

### 2. init
- initialize virtual enviroments 
- install request packages (requirement of the projects) 
- create pipfile.lock (produce determinated build)


pipenv install requests

### 3. activate the enviroment
pipenv shell

### 4. install project requirements
pip freeze > requirements.txt
<br>**OR**<br>
pipenv install -r ./requirements.txt

# Useful commands

________________________________________

### Developer enviroment
#### Install packages (e.g. pytest) only for development enviroment
pip install pytest --dev
pip install <package> --dev
    
________________________________________
### Push your code to deployment
pipenv lock

________________________________________
### Install dependencies from the pipfile.lock
pipenv install --ignore-pipfile

________________________________________

### Display project dependencies 
pipenv lock -r

________________________________________
### Unistall packages
pipenv unistall <package> <br> **OR** <br>
pipenv unistall pytest
________________________________________
### Run a command inside the enviroment
pipenv run python my_script.py

________________________________________
### Deactivate the enviroment
exit

________________________________________
### Change the python interpreter version
pipenv --python 3.6

________________________________________
### Remove the enviroment that we are using
pipenv --rm

________________________________________
### Path to our current enviroment
pipenv --venv

________________________________________
### Display graph dependencies
pipenv graph
