```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f=[i for i in range(0,10)]
g=[i for i in range(0,20,2)]

fig = plt.figure() 
ay=fig.add_subplot(111)
#plt.scatter(f,g,color='r')
ax = fig.add_subplot(111) # x,y,z #111= nella stessa img
ax.plot(f, g, color='lightblue', linewidth=2)
ay.scatter(f,g,color='r')
ax.set(xlim=[0,10],ylim=[0,25])
plt.grid()
```

### LINEAR MODEL

```python
df=pd.DataFrame({'Attack':[1,3,4,5,8,2,3,4],'Defense':[5,3,3,5,6,2,3,5]})
sns.lmplot(x='Attack', y='Defense', data=df, fit_reg=True)   # Color by evolution stage) # LINEAR MODEL 
# Alternative way
# sns.lmplot(x=df.Attack, y=df.Defense)
```

```python
df=pd.DataFrame({'Attack':[1,3,4,5,8,2,3,4],'Defense':[5,3,3,5,6,2,3,5],'Stage':[1,2,2,3,3,3,1,2]})
sns.lmplot(x='Attack', y='Defense', data=df, fit_reg=True, hue='Stage')   # Color by evolution stage) # LINEAR MODEL 
```

```python
df=pd.DataFrame({'Attack':[1,3,4,5,8,2,3,4],'Defense':[5,3,3,5,6,2,3,5],'Stage':[1,2,2,3,3,3,1,2],'S':['a','a','a','a','b','b','b','b']})

sns.pairplot(df, hue='S', markers=["o","p"],x_vars=["Attack","Defense"],y_vars=["Attack","Defense"])
```

```python
df = df.drop(['Stage'], axis=1)
sns.boxplot(data=df)
```

ELIMINA COLONNE

```python
#COLORI
pkmn_type_colors =  ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                    ]
                   
```

```python
sns.set_style('whitegrid')
sns.violinplot(x='Defense', y='Attack', data=df,palette=pkmn_type_colors)
```

```python
sns.swarmplot(x='Defense', y='Attack', data=df, 
              palette=pkmn_type_colors)
```

2 FIGURE ASSIEME

```python
plt.figure(figsize=(10,6))
sns.swarmplot(x='Defense', y='Attack', data=df, 
              palette=pkmn_type_colors)
sns.violinplot(x='Defense', y='Attack', data=df,palette=pkmn_type_colors)
```

```python
corr = df.corr()
# Heatmap
sns.heatmap(corr)
```
